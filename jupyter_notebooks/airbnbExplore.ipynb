```python
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport

pd.set_option('display.max_columns', None)
```

```python
def clean_listings(df):
    str_cols = ['bathrooms_text', 'price', 'room_type', 'property_type', 'neighbourhood_cleansed', 'name']
    irrelevant_cols = ['scrape_id', 'neighbourhood_group_cleansed', 'calendar_updated', 'license', 'bathrooms']
    
    numeric_df = df.copy().select_dtypes(include=np.number)
    numeric_df[str_cols] = df[str_cols]
    filtered_df = numeric_df.drop(columns=irrelevant_cols, axis=1, errors='ignore')
    
    return filtered_df[filtered_df.columns[::-1]]
```

```python
# change paths to csv files on your local machine
# download large listings.csv.gz files
amf_df = clean_listings(pd.read_csv('amsterdam_data/listings_1.csv'))
atf_df = clean_listings(pd.read_csv('athens_data/listings.csv'))
```
